# A. Build a baseline model 

```python
import pandas as pd
import numpy as np
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data
```

```python
concrete_data.shape
```

```python
concrete_data.describe()
```

```python
concrete_data.isnull().sum()
#our output shows that our data is very clean and no columns need to be dropped or anything. 
```

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
predictors.head()
```

```python
target.head()
```

```python
n_cols = predictors.shape[1] # number of predictors
n_cols
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
# create model
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
model.add(Dense(1))
    
# compile model
model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()
```

```python
list_of_mean_squared_errors_A=[]

a=50

while a>0:
    a=a-1
    pred_train,pred_test,target_train,target_test=train_test_split(predictors,target,test_size=0.3)
    # fit the model
    model.fit(pred_train,target_train ,epochs=50, verbose=2)
    prediction=model.predict(pred_test)
    list_of_mean_squared_errors_A.append(mean_squared_error(prediction,target_test))
    

```

```python
import statistics 
a=statistics.mean(list_of_mean_squared_errors_A)
a
```

```python
b=statistics.stdev(list_of_mean_squared_errors_A)
b
```

```python
ll=[]
l=[]
l.append('A')
l.append(a)
l.append(b)
ll.append(l)
ll
```

# B Normalize the data 

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# create model
model1 = Sequential()
model1.add(Dense(10, activation='relu', input_shape=(n_cols,)))
model1.add(Dense(1))
    
# compile model
model1.compile(optimizer='adam', loss='mean_squared_error')
model1.summary()
```

```python
#normalizing the predictors
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
list_of_mean_squared_errors_B=[]

a=50

while a>0:
    a=a-1
    pred_train,pred_test,target_train,target_test=train_test_split(predictors_norm,target,test_size=0.3)
    # fit the model
    model1.fit(pred_train,target_train ,epochs=50, verbose=2)
    prediction=model1.predict(pred_test)
    list_of_mean_squared_errors_B.append(mean_squared_error(prediction,target_test))
    

```

```python
import statistics 
a=statistics.mean(list_of_mean_squared_errors_B)
a
```

```python
b=statistics.stdev(list_of_mean_squared_errors_B)
b
```

```python
l=[]
l.append('B')
l.append(a)
l.append(b)
ll.append(l)
ll
```

# D Increase the number of hidden layers 

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# create model
model3 = Sequential()
model3.add(Dense(10, activation='relu', input_shape=(n_cols,)))
model3.add(Dense(10, activation='relu'))
model3.add(Dense(10, activation='relu'))
model3.add(Dense(10, activation='relu'))
model3.add(Dense(1))
    
# compile model
model3.compile(optimizer='adam', loss='mean_squared_error')
model3.summary()
```

```python
#normalizing the predictors
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
list_of_mean_squared_errors_D=[]

a=50

while a>0:
    a=a-1
    pred_train,pred_test,target_train,target_test=train_test_split(predictors_norm,target,test_size=0.3)
    # fit the model
    model3.fit(pred_train,target_train ,epochs=50, verbose=2)
    prediction=model3.predict(pred_test)
    list_of_mean_squared_errors_D.append(mean_squared_error(prediction,target_test))
    
```

```python
import statistics 
a=statistics.mean(list_of_mean_squared_errors_D)
a
```

```python
b=statistics.stdev(list_of_mean_squared_errors_D)
b
```

```python
l=[]
l.append('D')
l.append(a)
l.append(b)
ll.append(l)
ll
```
